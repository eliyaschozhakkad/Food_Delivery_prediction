```python
import pandas as pd
```

### Model Training

```python
#Reading the data
df = pd.read_csv('data/zomato.csv')
df.head()
```

```python
#Removing unwanted columns
df.drop(columns=['ID','Delivery_person_ID','Order_Date','Type_of_order'],inplace = True)
df.head()
```

```python
#Splitting into independent and dependent features
X = df.iloc[:,:-1]
y = df[['Time_taken (min)']]
```

```python
X.head()
```

```python
cols_to_clean = ['Restaurant_latitude','Restaurant_longitude','Delivery_location_latitude','Delivery_location_longitude']
for col in cols_to_clean:
    X[col] = abs(X[col])
```

```python

from geopy.distance import geodesic
```

```python
#Calculation of distance from restaurent to destination
X['Distance'] = X.apply(lambda row: geodesic((row['Restaurant_latitude'], row['Restaurant_longitude']), (row['Delivery_location_latitude'], row['Delivery_location_longitude'])).miles, axis=1)
```

```python
X.head()
```

```python
#Remove unwanted columns
X.drop(columns = ['Restaurant_latitude','Restaurant_longitude','Delivery_location_latitude','Delivery_location_longitude','Time_Orderd','Time_Order_picked'],inplace = True)
```

```python
X.head()
```

```python
X.info()
```

```python
#Numerical and Categorical featuers
categorical_features = X.select_dtypes(include='object').columns
numerical_features = X.select_dtypes(exclude='object').columns
```

```python
X[numerical_features]
```

```python
X[categorical_features]
```

```python
# Define the custom ranking for each ordinal variable
```

```python
weather_categories = ['Sunny','Windy','Stormy','Sandstorms','Fog','Cloudy']
traffic_categories = ['Low','Medium','High','Jam']
vehicletype_categories = ['scooter','electric_scooter','motorcycle','bicycle']
festival_map = ['No','Yes']
city_categories = ['Urban','Metropolitian','Semi-Urban'] 
```

```python
from sklearn.impute import SimpleImputer #Handling Missing values
from sklearn.preprocessing import StandardScaler #Handle Feature Scaling
from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding

##pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
##Numerical Pipeline

num_pipeline = Pipeline(
    steps = [
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
    ]
    
)

##Categorical Pipeline

cat_pipeline = Pipeline(
    steps = [
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('ordinalencoder',OrdinalEncoder(
            categories=['weather_categories','traffic_categories','vehicletype_categories','festival_map','city_categories'])),
        ('scaler',StandardScaler())
    ]

)

preprocessor = ColumnTransformer([
    ('num_pipeline',num_pipeline,numerical_features),
    ('cat_pipeline',cat_pipeline,categorical_features)
])
```

```python
##Numerical Pipeline

num_pipeline = Pipeline(
    steps = [
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
    ]
    
)

##Categorical Pipeline

cat_pipeline = Pipeline(
    steps = [
        ('imputer',SimpleImputer(strategy='most_frequent')),
         ('ordinalencoder', OrdinalEncoder(categories=[
            ['Sunny','Windy','Stormy','Sandstorms','Fog','Cloudy'],
            ['Low','Medium','High','Jam'],
            ['scooter','electric_scooter','motorcycle','bicycle'],
            ['No','Yes'],
            ['Urban','Metropolitian','Semi-Urban']
        ])),
        ('scaler',StandardScaler())
    ]

)

preprocessor = ColumnTransformer([
    ('num_pipeline',num_pipeline,numerical_features),
    ('cat_pipeline',cat_pipeline,categorical_features)
])
```

```python
## Train test split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=30)
```

```python
X_train
```

```python
X_train = pd.DataFrame(preprocessor.fit_transform(X_train),columns = preprocessor.get_feature_names_out())
X_test = pd.DataFrame(preprocessor.transform(X_test),columns = preprocessor.get_feature_names_out())
```

```python
X_train
```

```python
X_train.head()
```

```python
## Model Training
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
import numpy as np
def evaluate_model(true,predicted):
    mae = mean_absolute_error(true,predicted)
    mse = mean_squared_error(true,predicted)
    rmse = np.sqrt(mse)
    r2_score1 = r2_score(true,predicted)

    return mae,rmse,r2_score1
```

```python
## Train multiple models

models = {
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'ElasticNet':ElasticNet()
}

model_list = []
r2_list = []

for i in range(len(models)):
    model = list(models.values())[i]
    model.fit(X_train,y_train)

    #Make Predictions
    y_pred = model.predict(X_test)

    mae,rmse,r2_score1=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print('RMSE',rmse)
    print('MAE',mae)
    print('r2 score',r2_score1*100)

    r2_list.append(r2_score1)
    print("="*35)
    print('\n')
```
